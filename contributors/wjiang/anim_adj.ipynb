```python
import ecco_v4_py as e4p
import xarray as xr
```

```python
import cartopy.crs as ccrs
projection = ccrs.LambertConformal(central_longitude = -55)
import cmocean
cmap = cmocean.cm.balance
import cartopy.feature as cfeature
```

```python
grid = xr.open_dataset('/efs_ecco/wjiang/data_folder/GRID_GEOMETRY_ECCO_V4r4_native_llc0090.nc')
xx = np.array(grid.XC).ravel()
yy = np.array(grid.YC).ravel()
```

```python
import matplotlib.animation as animation
```

```python
for name in ['empmr', 'pload', 'qnet', 'qsw', 'saltflux', 'spflx', 'tauu', 'tauv']:
    sens = e4p.llc_compact_to_tiles(globals_dict['return_vars']['adxx_'+name])

    vmax = np.max(np.abs(sens[77]))*0.8
    print(name,vmax)
    plt.clf()
    fig = plt.figure()
    ax = plt.axes(projection = projection)
    it = 0
    pcm2 = plt.pcolormesh(grid.XC[2],grid.YC[2],sens[it,2],transform = ccrs.PlateCarree(),vmax = vmax,vmin = -vmax,cmap = cmap)
    pcm6 = plt.pcolormesh(grid.XC[6],grid.YC[6],sens[it,6],transform = ccrs.PlateCarree(),vmax = vmax,vmin = -vmax,cmap = cmap)
    pcmx = plt.pcolormesh(grid.XC[10],grid.YC[10],sens[it,10],transform = ccrs.PlateCarree(),vmax = vmax,vmin = -vmax,cmap = cmap)
    plt.title(name+f' lag = {it+1} week')
    # ax.tricontourf(xx,yy,varexp[2].ravel(),levels = 50,transform = ccrs.PlateCarree())
    plt.colorbar(extend = 'both')
    ax.set_extent((-80,-10,15,90))
    ax.add_feature(cfeature.LAND,facecolor = 'grey',zorder = 10)
    plt.tight_layout()
    
    def update(frame):
        pcm2.set_array(sens[frame,2].ravel())
        pcm6.set_array(sens[frame,6].ravel())
        pcmx.set_array(sens[frame,10].ravel())
        ax.set_title(name+f' lag = {frame+1} week')
        return pcm2, pcm6, pcmx
    
    ani = animation.FuncAnimation(
        fig, update, frames=78, blit=True, interval=100
    )
    plt.show()
    
    ani.save(name+'_adj.mp4')
```
